```python
INHERITANCE
SINGLE INHERITANCE anables a derived class to inherite the properties of single parent class does anabling code reusablility and addition of  
new features to existing program
```

```python
class parent:
    def a(self):
        print("we are in parent class")
class child(parent):
    def c(self):
        print("we are in child class")
obj2=child()
obj2.c()
obj2.a()
```

```python
MULTIPLE INHERITANCE 
when a class can be derived from more than 1 base class this type of inheritance is called multyple inheritance
     A        B
          C
```

```python
class parent1:
    def a(self):
        print("we are in parent1 class")
class parent2:
    def b(self):
        print("we are in parent2 class")
class child(parent1,parent2):
    def c(self):
        print("we are in child class")
obj2=child()
obj2.c()
obj2.a()
obj2.b()
```

```python
MULTY LEVEL INHERITANCE
in multy level inheritance features of base class and the derived class are further inherited into new derived class this is simillar to a relationship 
representing a child and a grandfather
         grandfather                           A
           father                              B
           child                               C
```

```python
class grandfather:
    def grand(self):
        print("i am grandfather")
class father(grandfather):
    def dad(self):
        print("i am father")
class child(father):
    def son(self):
        print("i am child")
obj2=child()
obj2.grand()
obj2.dad()
obj2.son()
```

```python
HIERARCHICAL INHERITANCE
when more than 1 derived class are created from 1 base class this type of inheritance is called as hierarchical inheritance 
                  A
            B   C   D   E
```

```python
class base:
    def b(self):
        print("i am base class")
class derived1(base):
    def d1(self):
        print("i am derived class 1")
class derived2(base):
    def d2(self):
        print("i am derived class 2")
class derived3(base):
    def d3(self):
        print("i am derived class 3")
obj1=derived1()
obj2=derived2()
obj3=derived3()
obj1.b()
obj1.d1()
obj2.d2()
obj3.d3()
```

```python
HYBRID INHERITANCE
inhertance consisting of muliple types of inheritance is known as hybrid inheritance
                               A
                           B       C
                               D
```

```python
class base:
    def b(self):
        print("i am base class")
class derived1(base):
    def d1(self):
        print("i am derived class 1")
class derived2(base):
    def d2(self):
        print("i am derived class 2")
class derived3(base):
    def d3(self):
        print("i am derived class 3")
class parent1:
    def a(self):
        print("we are in parent1 class")
class parent2:
    def b1(self):
        print("we are in parent2 class")
class child(derived3,parent1,parent2):
    def c(self):
        print("we are in child class")
o1=derived1()
o2=derived2()
o3=derived3()
o1.b()
o1.d1()
o2.d2()
o3.d3()
obj2=child()
obj2.c()
obj2.a()
obj2.b1()
```

```python
when ever the function name is declared same in both the classes (child,base)which is having different types of parameters or different types of data
types is known as METHOD OVERLOADING
when both the inherited classes (base and child)are having same function name and same number of parameters are present in both the base andchild class
together in this condition the child class method will over write the base class method.if you are calling the function to print the data it will only 
print the over writed method data (child class method dta)
```

```python
class parent():
    def show(self):
        print("inside parent")
class child():
    def show(self):
        print("inside child")
class grandchild():
    def show(self):
        print("inside grandchild")
g=grandchild()
g.show()
g.show()
```
